```python
# Resolução do exercício 1
n = int(input('Digite um número maior que 1, e irei te dizer todos os números primos antes dele: '))
 
if n > 1: #Colocando a condição para que o número seja maior que 1
    for i in range(2,n + 1):
            x, mult = 2, 0
            while(x<i):
                if(i % x == 0): #Se não houver resto, o número é primo
                    mult += 1
                x+=1
            else:
                if(mult==0):
                    print("\nO número {} é primo.".format(i))
else:
    print('\n')
    print('Você é um belo de um pilantra!!')
```

### Ex 2:

```python
#########################################
############## Dicionários ##############
#########################################

 

menu = { 'Cheese burger': 10.0, 
        'Bacon burger': 13.0, 
        'Vegan burger': 10.0, 
        'Refrigerante': 5.0, 
        'Batata frita': 6.0, 
        'Batata frita com Bacon e Cheddar': 8.0, 
        'Sorvete': 2.0, 
        'Sundae': 6.0 }

 

menu2 = {1:{'Cheese burger': 10.0}, 
        2:{'Bacon burger': 13.0}, 
        3:{'Vegan burger': 10.0}, 
        4:{'Refrigerante': 5.0}, 
        5:{'Batata frita': 6.0}, 
        6:{'Batata frita + Bacon&Cheddar': 8.0}, 
        7:{'Sorvete': 2.0}, 
        8:{'Sundae': 6.0 },
        9:'Finlizar Pedido'}

 

####################################
############## Código ##############
####################################

 
 
############## Print Menu #############

print('Menu somos todos amigos!\n')
count = 0
for item,preco in menu.items():
    count += 1
    print('  ({}) {} = R${}'.format(count,item,preco))
print(' (9) Finalizar Pedido')

 

########### Criando o pedido ###########
list_prduto = []
list_val_uni = []
list_qtd = []
list_sub_total = []
try:
    while True:
        num = int(input('\nSegundo o menu acima,\n Digite o número do seu pedido: '))

 

        if num != 9:
            qtd = int(input('Quantidade deste item em específico? '))
            if num == 1:
                alimento = 'Cheese burger'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 2:
                alimento = 'Bacon burger'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 3:
                alimento = 'Vegan burger'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 4:
                alimento = 'Refrigerante'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 5:
                alimento = 'Batata frita'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 6:
                alimento = 'Batata frita + Bacon&Cheddar'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 7:
                alimento = 'Sorvete'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num == 8:
                alimento = 'Sundae'
                print('Você solicitou {} {}(s)'.format(qtd,alimento))
                list_prduto.append(alimento)
                list_val_uni.append(menu2[num][alimento])
                list_qtd.append(qtd)
                list_sub_total.append(qtd*menu2[num][alimento])
            elif num > 9:
                print('Opss, este não foi um produto valido.')
        else:
            break

 
    print(color.BOLD)
    pedido_final = pd.DataFrame(index=list_prduto)
    #pedido_final['Produto'] = list_prduto
    pedido_final['Valor Unitário (R$)'] = list_val_uni
    pedido_final['Quantidade'] = list_qtd
    pedido_final['Sub Total (R$)'] = list_sub_total
    print(41*'__')
    print('\n',pedido_final)
    print(41*'__')
    print('\nValor Total: R$',sum(list_sub_total))
    input('Observações no pedido: ')
    print('\nPedido Realizado com sucesso! Muito obrigado pela preferência.')
except ValueError:
    print('Opss, o caracter digitado não foi um produto válido!!')
```
