```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv("exemplos_limpeza_dados.csv")
```

```python
df.info()
```

```python
df
```

```python
#remover valores duplicados
df.drop_duplicates(inplace=True)
```

```python
#Preenchimento de dados faltantes
df['Age'].fillna(df['Age'].mean(),inplace=True) #Preencher com a média
df['Salary'].fillna(df['Salary'].median(),inplace=True) #Preencher com a mediana
```

```python
# Tratando dados categóricos
le = LabelEncoder()
df["Grade_encoder"] = le.fit_transform(df["Grade"])
df['Purchased_encoder'] = le.fit_transform(df['Purchased'])
```

```python
df
```

```python
# Usando o OneHotEncoder padrão do pandas
df2 = pd.get_dummies(df['Country'])
df2
```

```python
df3 = pd.concat([df,df2], axis=1)
df3
```

```python
# Filtrar colunas relevantes
df3.drop(columns = ['Country','Grade','Purchased'], inplace = True)
```

```python
df3
```

```python
#Separando treino e teste
X = df3.drop(columns = ['Purchased_encoder'])
y = df3['Purchased_encoder']
x_train,x_test,y_train,y_test =  train_test_split(X,y, test_size=0.2)
```

```python
#Normalizar dados numéricos
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```
