```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Line Plot

x = [1,2,3]
xx = [1.5,2.5,3.4]
y = [1000,2000,3000]
yy = [589,1500,3400]
yyy = [2200,1200,1650]

plt.figure(figsize=(10,6))
plt.plot(x,y, '-ro',label='CDB') # '-go' estilo do grafico linha(-),red(g),bolinha(o)
plt.plot(x,yy,':gx',label='LCI')
plt.plot(xx,yy,'--b^',label='Acoes')
plt.plot(xx,yyy,'k--',label='Fundos')

plt.axis((1,4,1000,3500))
plt.xlabel("Tempo (Anos)",size = 15)
plt.ylabel("Montante (R$)",size = 15)
plt.title("R$ x Anos",size = 25)
plt.legend()

plt.show()
```

```python
# Barplot por faixas

Idade = ['16-20','21-26','27-35','36-45','+45']
consumo = [295,1000,670,800,1230]

plt.figure(figsize=(12,6))
plt.bar(Idade,consumo,color='purple')#,labels=Idade)

plt.xlabel("Faixa etaria",size = 15)
plt.ylabel("Pessoas com mais de 1m (R$)",size = 15)
plt.title("Milhonarios por faixa etaria",size = 25)

plt.show()
```
