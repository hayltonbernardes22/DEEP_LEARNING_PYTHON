### Para se calcular a distancia entre os pontos deve se calcular a distancia euclidiana dos pontos mais proximos ao dado a ser analisado

```python
# Distancia Euclidiana
import math
def dist_euclidiana(amostra1,amostra2):
    soma = 0
    if len(amostra1) == len(amostra2):
        for i in range(len(amostra1)):
            soma += ((amostra1[i] - amostra2[i]) ** 2)
        return math.sqrt(soma)
    else:
        return print('Listas com valores distintos')
```

```python
# Exemplo de distancias entre pontos de uma lista
x = [3,10,25,23]
y = [4,8,16,23]
dist_euclidiana(x,y)
```
